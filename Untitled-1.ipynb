```python
mi-app-medicina/
‚îÇ
‚îú‚îÄ‚îÄ index.html       ‚Üê Estructura visual
‚îú‚îÄ‚îÄ style.css        ‚Üê Dise√±o y colores
‚îú‚îÄ‚îÄ app.js           ‚Üê L√≥gica matem√°tica
‚îú‚îÄ‚îÄ lib/
‚îÇ   ‚îî‚îÄ‚îÄ math.js      ‚Üê Biblioteca para c√°lculos

<!DOCTYPE html>
<html lang="es">
<head>
  <meta charset="UTF-8">
  <title>App M√©dica de Matem√°ticas</title>
  <link rel="stylesheet" href="style.css">
</head>
<body>
  <h1>Calculadora M√©dica</h1>
  <input id="input" placeholder="Ej: 500mg / 8h por 3 d√≠as">
  <button onclick="resolver()">Calcular</button>
  <p id="resultado"></p>

  <script src="lib/math.js"></script>
  <script src="app.js"></script>
</body>
</html>
function resolver() {
  const entrada = document.getElementById('input').value;

  // Aqu√≠ interpretamos el c√°lculo (por ahora b√°sico)
  try {
    const resultado = math.evaluate(entrada);
    document.getElementById('resultado').innerText = "Resultado: " + resultado;
  } catch (error) {
    document.getElementById('resultado').innerText = "Error: expresi√≥n inv√°lida";
  }
}
body {
  font-family: sans-serif;
  background: #f0f4f8;
  text-align: center;
  padding: 2em;
}

input {
  width: 300px;
  padding: 10px;
  font-size: 1em;
}

button {
  padding: 10px 20px;
  font-size: 1em;
  margin-left: 10px;
}

#resultado {
  margin-top: 20px;
  font-weight: bold;
}
<!DOCTYPE html>
<html lang="es">
<head>
  <meta charset="UTF-8">
  <title>Calculadora M√©dica</title>
  <link rel="stylesheet" href="style.css">
</head>
<body>
  <h1>üß† Calculadora M√©dica</h1>

  <div class="card">
    <h2>Dosis por peso</h2>
    <input id="peso" type="number" placeholder="Peso en kg">
    <input id="dosis" type="number" placeholder="Dosis mg/kg">
    <button onclick="calcularDosis()">Calcular dosis total</button>
    <p id="resultadoDosis"></p>
  </div>

  <div class="card">
    <h2>IMC</h2>
    <input id="pesoIMC" type="number" placeholder="Peso en kg">
    <input id="estatura" type="number" placeholder="Estatura en metros">
    <button onclick="calcularIMC()">Calcular IMC</button>
    <p id="resultadoIMC"></p>
  </div>

  <div class="card">
    <h2>Regla de Clark (Ni√±os)</h2>
    <input id="pesoNi√±o" type="number" placeholder="Peso en kg">
    <input id="dosisAdulto" type="number" placeholder="Dosis adulta (mg)">
    <button onclick="reglaClark()">Calcular dosis infantil</button>
    <p id="resultadoClark"></p>
  </div>

  <script src="app.js"></script>
</body>
</html>
function calcularDosis() {
  const peso = parseFloat(document.getElementById("peso").value);
  const dosis = parseFloat(document.getElementById("dosis").value);

  if (isNaN(peso) || isNaN(dosis)) {
    document.getElementById("resultadoDosis").innerText = "Completa los campos correctamente.";
    return;
  }

  const total = peso * dosis;
  document.getElementById("resultadoDosis").innerText = `Dosis total: ${total.toFixed(2)} mg`;
}

function calcularIMC() {
  const peso = parseFloat(document.getElementById("pesoIMC").value);
  const estatura = parseFloat(document.getElementById("estatura").value);

  if (isNaN(peso) || isNaN(estatura) || estatura === 0) {
    document.getElementById("resultadoIMC").innerText = "Datos inv√°lidos.";
    return;
  }

  const imc = peso / (estatura * estatura);
  document.getElementById("resultadoIMC").innerText = `IMC: ${imc.toFixed(2)} (${clasificarIMC(imc)})`;
}

function clasificarIMC(imc) {
  if (imc < 18.5) return "Bajo peso";
  if (imc < 25) return "Normal";
  if (imc < 30) return "Sobrepeso";
  return "Obesidad";
}

function reglaClark() {
  const peso = parseFloat(document.getElementById("pesoNi√±o").value);
  const dosisAdulto = parseFloat(document.getElementById("dosisAdulto").value);

  if (isNaN(peso) || isNaN(dosisAdulto)) {
    document.getElementById("resultadoClark").innerText = "Completa todos los campos.";
    return;
  }

  const dosisNi√±o = (peso * dosisAdulto) / 70; // Suponemos adulto promedio de 70kg
  document.getElementById("resultadoClark").innerText = `Dosis infantil: ${dosisNi√±o.toFixed(2)} mg`;
}
body {
  font-family: Arial, sans-serif;
  background-color: #eef3f7;
  padding: 20px;
  text-align: center;
}

.card {
  background: white;
  border-radius: 10px;
  padding: 20px;
  margin: 20px auto;
  max-width: 400px;
  box-shadow: 0 0 10px rgba(0,0,0,0.1);
}

input {
  display: block;
  margin: 10px auto;
  padding: 10px;
  width: 80%;
  font-size: 16px;
}

button {
  padding: 10px 20px;
  font-size: 16px;
  background: #0077cc;
  color: white;
  border: none;
  border-radius: 5px;
  cursor: pointer;
}

button:hover {
  background: #005fa3;
}

p {
  margin-top: 10px;
  font-weight: bold;
}
<div class="card">
  <h2>Infusi√≥n IV (ml/h)</h2>
  <input id="volumen" type="number" placeholder="Volumen en ml">
  <input id="tiempo" type="number" placeholder="Tiempo en horas">
  <button onclick="calcularInfusion()">Calcular</button>
  <p id="resultadoInfusion"></p>
</div>

<div class="card">
  <h2>Conversi√≥n mg ‚Üî mL</h2>
  <input id="dosisDeseada" type="number" placeholder="Dosis deseada (mg)">
  <input id="dosisDisponible" type="number" placeholder="Dosis disponible (mg)">
  <input id="volumenDisponible" type="number" placeholder="Volumen disponible (mL)">
  <button onclick="convertirMgML()">Calcular volumen a administrar</button>
  <p id="resultadoConversion"></p>
</div>
function calcularInfusion() {
  const volumen = parseFloat(document.getElementById("volumen").value);
  const tiempo = parseFloat(document.getElementById("tiempo").value);

  if (isNaN(volumen) || isNaN(tiempo) || tiempo === 0) {
    document.getElementById("resultadoInfusion").innerText = "Datos inv√°lidos.";
    return;
  }

  const velocidad = volumen / tiempo;
  document.getElementById("resultadoInfusion").innerText = `Velocidad: ${velocidad.toFixed(2)} ml/h`;
}

function convertirMgML() {
  const dosisDeseada = parseFloat(document.getElementById("dosisDeseada").value);
  const dosisDisponible = parseFloat(document.getElementById("dosisDisponible").value);
  const volumenDisponible = parseFloat(document.getElementById("volumenDisponible").value);

  if (isNaN(dosisDeseada) || isNaN(dosisDisponible) || isNaN(volumenDisponible) || dosisDisponible === 0) {
    document.getElementById("resultadoConversion").innerText = "Datos inv√°lidos.";
    return;
  }

  const volumenAdministrar = (dosisDeseada * volumenDisponible) / dosisDisponible;
  document.getElementById("resultadoConversion").innerText = `Debes administrar: ${volumenAdministrar.toFixed(2)} mL`;
}
```
